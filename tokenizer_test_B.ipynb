```python
from nltk import FreqDist
from nltk.tag import pos_tag

stopwords = ["Don", "of", "and", "t", "I", "m"]
ipod_tokens = pos_tag(retokenize.tokenize(ipod_all))
```

```python
ipod_tokens
```

```python
names_list = [t[0] for t in ipod_tokens if t[1] == "NN" and t[0] not in stopwords]
fd_names = FreqDist(names_list)
```

```python
fd_names.N()
```

```python
fd_names['iPod']
```

```python
fd_names.freq("iPod")
```

```python
fd_names.most_common(5)
```

```python
for line in open('ipodwords.txt'):
    print(line.strip())
```

```python
ipodwords = set(line.strip() for line in open('ipodwords.txt')) # 익명함수, 클러저! 
```

```python
ipodwords.union(set(['iPod', 'time', 'apple']))
```

```python
ipodcount = {}
for ipod in ipod_all.lower().split():
    ipod = ipod.replace(".", "")
    ipod = ipod.replace(",", "")
    ipod = ipod.replace(":", "")
    ipod = ipod.replace("\"", "")
    ipod = ipod.replace("*", "")
    
    if ipod not in ipodwords:
        if ipod not in ipodcount:
            ipodcount[ipod] = 1
        else:
            ipodcount[ipod] += 1
```

```python
ipodcount
```

```python
import pandas as pd
```

```python
i_print = int(input('how many list >> '))
fd_names.most_common(i_print)
```

```python
ipod_list = fd_names.most_common(i_print)
df = pd.DataFrame(ipod_list, columns=['Word', 'Count'])
df
```

```python
df.plot.bar(x = 'Word', y = 'Count')
```
