```python
# !pip install konlpy
```

```python
from konlpy.tag import Okt
```

```python
okt = Okt()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(tokenizer = okt.morphs, ngram_range=(1, 2), min_df = 3, max_df = 0.9)
tfidf
```

```python
tfidf.fit(train_df['document'])
```

```python
train_tfidf_df = tfidf.transform(train_df['document'])
train_tfidf_df
```

```python
tfidf.inverse_transform(train_tfidf_df)
```

```python
train_tfidf_df[0].toarray()
```

```python
# 테스트하기 위해서 테스트 데이터도 (1) fit, transform (2) transform  => (2)번이 정답 
# 테스트데이터를 tfidx vectorize할 때는 fit한 후, transform 시켜야 함.
test_tfidf_df = tfidf.transform(test_df['document'])
test_tfidf_df
```

```python
## 1. 로지스틱 회귀 객체 생성
## 2. 트레인데이터로 학습
## 3. 테스트데이터로 예측
## 4. 정확도를 구해보세요
## 5. 입력한 데이터로 예측하기 
### (입력한 데이터 한글이외 제외시키고, transform한 후 예측)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state = 0)
```

```python
lr.fit(train_tfidf_df, train_df['label'])
```

```python
test_predict = lr.predict(test_tfidf_df)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('감성 분석 정확도 : ', round(accuracy_score(test_df['label'], test_predict), 3))
```
