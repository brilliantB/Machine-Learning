```python
import json

# xml -> <money>1000</money>
# eXtensible Markup Langeuages(확장가능한 XML)
# json -> {'money' : 1000}
# JavaScript Object Notation(JSON)
```

```python
file_name = '/content/sample_data/etnews.kr_facebook_2016-01-01_2018-08-01_4차 산업혁명.json'
json_data = json.loads(open(file_name, 'r', encoding='utf-8').read())
json_data
```

```python
all_data = ""
```

# 새 섹션

```python
import re
```

```python
for item in json_data: #dic의 리스트:
  if 'message' in item.keys():
    all_data = all_data + re.sub(r'[^\w]', ' ', item['message']) + " "
print(all_data)
```

```python
# 명사 단어 추출!
# 카운트 딕셔너리 확인! 
# 5~10위 프린트
```

```python
noun_list3 = okt.nouns(all_data)
noun_list3
```

```python
count3 = Counter(noun_list3)
count3
```

```python
count3.most_common(5)
```

```python
word_count2 = dict()
for tag, count3 in count3.most_common(15):
    if (len(tag)>1):
        word_count2[tag] = count3
print(word_count2)
```
