```python
# Label encoding
from sklearn.preprocessing import LabelEncoder
```

```python
import numpy as np
import pandas as pd
```

```python
data = pd.Series(['apple', 'summer', 'spring'])
data
```

```python
encoder = LabelEncoder()
encoder.fit(data)        # 시리즈의 값들이 카테고리 성격을 가진다.
```

```python
target = pd.Series(['apple', 'summer', 'summer', 'spring', 'spring'])
target
```

```python
encoder.transform(target)
```

```python
target2 = encoder.transform(target)
target2
```

```python
### 2개 컬럼 이상 라벨 인코딩하기 ###
```

```python
words = pd.Series(['apple', 'summer', 'summer', 'spring', 'spring'])
hobby = pd.Series(['book', 'run', 'talk', 'coffee', 'song'])
```

```python
df = pd.DataFrame({'words' : words, "hobby" : hobby})
df
```

```python
columns = df.columns
columns
```

```python
encoder2 = LabelEncoder()
```

```python
for col in columns:
    df[col] = encoder2.fit_transform(df[col])
```

```python
df
```

# (실습) 원핫+라벨인코딩, 라벨링해보기 
- one hot : ST
- label : bank name, city

```python
banklist = pd.read_csv('banklist.csv')
```

```python
banklist.head()
```

```python
banklist.info()
```

```python
st = banklist.ST
data = pd.get_dummies(st)
data
```

```python
data = banklist[['Bank Name', 'City']]
data
```

```python
columns = data.columns
columns
```

```python
encoder = LabelEncoder()
for col in columns:
    data[col] = encoder.fit_transform(data[col])
```
